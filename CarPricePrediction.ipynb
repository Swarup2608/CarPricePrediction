```python
# Importing Dependenicies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
df = pd.read_csv('car.csv')
df
```

```python
# checking the distribution of categorical data
print(df.Fuel_Type.value_counts())
print(df.Seller_Type.value_counts())
print(df.Transmission.value_counts())
```

```python
# encoding "Fuel_Type" Column
df.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)

# encoding "Seller_Type" Column
df.replace({'Seller_Type':{'Dealer':0,'Individual':1}},inplace=True)

# encoding "Transmission" Column
df.replace({'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
```

```python
df.head()
```

```python
# separating the data and label
X = df.drop(columns=['Car_Name','Selling_Price'],axis=1)
Y = df['Selling_Price']
```

```python
# Dividing to testing and training Data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state=2)
```

```python
# Linear Regression
model = LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
# R squared Error of training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = metrics.r2_score(X_train_prediction, Y_train)
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
# R squared Error of Testing Data
test_data_prediction = model.predict(X_test)
error_score = metrics.r2_score(Y_test, test_data_prediction)
print("R squared Error : ", error_score)
```

```python
# Predictive System
l = ['Year','Present Price','KMS','Fuel Type 0 or 1 or 2','Seller Type 0 or 1','Transmission 0 or 1','Owner 0 or 1']
Xnew = []
carname = input("Enter the Car Name : ")
for i in range(len(l)):
  Xnew.append(input("Enter the "+l[i]+": "))
# Converting to array
Xnewa = np.asarray(Xnew)
# Reshaping
Xres = Xnewa.reshape(1,-1)
# Prediction 
pred = model.predict(Xres)
print("Selling price of ",carname, pred)
```
